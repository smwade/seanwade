```python
from random import random
from math import log

from matplotlib import pyplot as plt
```

```python
rand = random
```

```python
class Distribution:

    def sample(self) -> float:
        raise NotImplemented
        
    def samples(self, n: int) -> list[float]:
        return [self.sample() for _ in range(n)]
```

```python
def plot_dist(dist, n=1000):
    samples = dist.samples(n)
    plt.hist(samples)
    plt.show()
```

```python
class UniformDist(Distribution):
    
    def __init__(self, a, b):
        self.a = a
        self.b = b
        
    def sample(self):
        return rand() * (self.b - self.a) + self.a
```

```python
plot_dist(UniformDist(2,7))
```

```python
class Binmomial(Distribution):

    def __init__(self, p, n):
        self.p = p
        self.n = n
        
    def sample(self):
        cnt = 0
        for _ in range(self.n):
            if rand() <= self.p:
                cnt += 1
        return cnt
```

```python
plot_dist(Binmomial(.4,100))
```

### Inverse Transform Method

```python
class ExponentialDist(Distribution):
    def __init__(self, lmbda):
        self.lmbda = lmbda
        
    def sample(self):
        return -(1/self.lmbda)*log(1-rand())
```
